```
import numpy as np
```

```
x = np.array([[18],[23],[28],[33],[38],[43],[48], [53],[58],[63]])
```

```
y = np.array([[871],[1132],[1042],[1356],[1488],[1638],[1569],[1754],[1866],[1900]])
```

```
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(x,y)
```

```
from sklearn.linear_model import LinearRegression
regressor = LinearRegression(n_jobs=1)
regressor.fit(x,y)
```

**Coeficientes** 

```
#b0
regressor.intercept_
```

```
 # b1
 regressor.coef_
```

**Previsão**

```
provisao1 = regressor.intercept_ + regressor.coef_*40
```

```
provisao1
```

```
provisao2 = regressor.predict([[40]])
provisao2
```

```
provisoes = regressor.predict(x)
provisoes
```

```
resultado = abs(y - provisoes).mean()
resultado
```

```
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```
mae = mean_absolute_error(y, provisoes)
mse = mean_squared_error(y, provisoes)
```

```
mae
```

```
mse
```

```
plt.plot(x, y, 'o')
plt.plot(x, provisoes,  color = 'red')
plt.title('Regressao linear simples')
plt.xlabel('Idade')
plt.ylabel('Custo')
```
