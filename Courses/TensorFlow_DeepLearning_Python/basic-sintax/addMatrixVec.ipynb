```
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```
a = tf.constant([9,8,7], name = 'a')
b = tf.constant([1,3,2], name = 'b')
```

```
soma = a + b
```

```
type(a)
```

```
print(a)
```

```
with tf.Session() as sess:
  print(sess.run(soma))
```
