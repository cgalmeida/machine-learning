```
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```
a = tf.constant([[1,2], [3,4]])
b = tf.constant([[-1,3], [4,2]])
```

```
multiplicacao = tf.matmul(a,b)
```

```
with tf.Session() as sess:
  print(sess.run(a))
  print('\n')
  print(sess.run(b))
  print('\n')
  print(sess.run(multiplicacao))
```

```
multiplicacao2 = tf.matmul(b,a)
```

```
with tf.Session() as sess:
  print(sess.run(multiplicacao2))
```

```
a2  = tf.constant([[2,3],[0,1],[-1,4]], name = 'a2')
b2  = tf.constant([[1,2,3],[-2,0,4]], name = 'b2')
```

```
multiplicacao3 =tf.matmul(a2,b2)
```

```
with tf.Session() as sess:
  print(sess.run(a2))
  print('\n')
  print(sess.run(b2))
  print('\n')
  print(sess.run(multiplicacao3))
```
